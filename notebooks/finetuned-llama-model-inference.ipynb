```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
# Load the model and tokenizer from the Hugging Face Hub
model = AutoModelForCausalLM.from_pretrained("saraHamza/Llama-2-7b-chat-vocab-grammer-finetune")
tokenizer = AutoTokenizer.from_pretrained("saraHamza/Llama-2-7b-chat-vocab-grammer-finetune")

```

# Example 1 **Word Meaning**

```python
# Perform inference
input_text = "what is the meaning of the word angry ? "
inputs = tokenizer(input_text, return_tensors="pt")
outputs = model.generate(
    **inputs,
    temperature=0.6,
    repetition_penalty=1.2,
    max_length=100,  # You can adjust this based on your needs
)

# Decode and print the generated text
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
```
